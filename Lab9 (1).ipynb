```python

import cv2
import numpy as np

# Load the original image
original_image = cv2.imread("brain.jpg")

# Function to add Gaussian noise to the image
def add_gaussian_noise(image, mean=0, std=5):
    gauss_noise = np.random.normal(mean, std, image.shape).astype(np.uint8)
    noisy_image = cv2.add(image, gauss_noise)
    return noisy_image

# Function to add salt-and-pepper noise to the image


# Add Gaussian noise to the image
noisy_image_gaussian = add_gaussian_noise(original_image)


cv2.imwrite("noisy_image_gaussian2.jpg", noisy_image_gaussian)


```

```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt

# Define file paths for the images
original_image_path = "brain.jpg"
noisy_image_path = "noisy_image_gaussian2.jpg"

# Load the original and noisy images
original_image = cv2.imread(original_image_path, cv2.IMREAD_GRAYSCALE)
noisy_image = cv2.imread(noisy_image_path, cv2.IMREAD_GRAYSCALE)

# Check if images were loaded successfully
if original_image is None:
    raise ValueError(f"Error loading original image from '{original_image_path}'.")
if noisy_image is None:
    raise ValueError(f"Error loading noisy image from '{noisy_image_path}'.")

# Continue with image processing and metrics calculation
# Calculate Mean Squared Error (MSE)
mse = np.mean((original_image - noisy_image) ** 2)

# Calculate Peak Signal-to-Noise Ratio (PSNR)
psnr = 10 * np.log10((255**2) / mse)

# Calculate Structural Similarity Index (SSIM)
ssim_score = ssim(original_image, noisy_image)

# Calculate and plot histograms
original_hist = cv2.calcHist([original_image], [0], None, [256], [0, 256])
noisy_hist = cv2.calcHist([noisy_image], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")

plt.subplot(2, 2, 2)
plt.imshow(noisy_image, cmap='gray')
plt.title("Noisy Image")

plt.subplot(2, 2, 3)
plt.plot(original_hist, color='b')
plt.title("Histogram of Original Image")

plt.subplot(2, 2, 4)
plt.plot(noisy_hist, color='r')
plt.title("Histogram of Noisy Image")

plt.tight_layout()

# Print and display the metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Peak Signal-to-Noise Ratio (PSNR): {psnr:.2f} dB")
print(f"Structural Similarity Index (SSIM): {ssim_score:.2f}")

plt.show()
```

```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt

# Load the original and noisy images
original_image = cv2.imread("brain.jpg", cv2.IMREAD_GRAYSCALE)
noisy_image = cv2.imread("noisy_image_gaussian2.jpg", cv2.IMREAD_GRAYSCALE)

# Reduce noise with a Gaussian blur
noisy_image_blurred = cv2.GaussianBlur(noisy_image, (5, 5), 0)

# Calculate Mean Squared Error (MSE)
mse = np.mean((original_image - noisy_image_blurred) ** 2)

# Calculate Peak Signal-to-Noise Ratio (PSNR)
psnr = 10 * np.log10((255**2) / mse)

# Calculate Structural Similarity Index (SSIM)
ssim_score = ssim(original_image, noisy_image_blurred)

# Calculate histograms
original_hist = cv2.calcHist([original_image], [0], None, [256], [0, 256])
blurred_hist = cv2.calcHist([noisy_image_blurred], [0], None, [256], [0, 256])

# Create a line graph to compare histograms
x = np.arange(256)
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")

plt.subplot(2, 2, 2)
plt.imshow(noisy_image_blurred, cmap='gray')
plt.title("Reduced Noise Image")

plt.subplot(2, 2, 3)
plt.plot(x, original_hist, color='b')
plt.title("Histogram of Original Image")

plt.subplot(2, 2, 4)
plt.plot(x, blurred_hist, color='r')
plt.title("Histogram of Reduced Noise Image")

plt.tight_layout()

# Create a line graph to compare histograms
plt.figure(figsize=(10, 4))

plt.plot(x, original_hist, color='b', label='Original Image')
plt.plot(x, blurred_hist, color='r', label='Reduced Noise Image')
plt.title("Histogram Comparison")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.legend()

# Print and display the metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Peak Signal-to-Noise Ratio (PSNR): {psnr:.2f} dB")
print(f"Structural Similarity Index (SSIM): {ssim_score:.2f}")

plt.show()
```
